```python


def getrandomstr(in_list,num):
    test_list = random.sample(in_list,num)
    test_str = ""
    for i in range(num):
        test_str = test_str+test_list[i]+','

    return test_str,test_list
```

```python
import copy
def get_train_str_list(in_list,test_list):
    out_str = ''
    train_list = copy.copy(in_list)
    for item in test_list:
        train_list.remove(item)
        
    for i in range(len(train_list)):
        out_str = out_str + train_list[i]+','
        
    return out_str,train_list
```

```python
#10% as test data
#10% as val data
import random
import copy
random.seed(0)


def getrandom_train_test(in_list,num):
    test_list = random.sample(in_list,num)
    train_list = copy.copy(in_list)
    for item in test_list:
        train_list.remove(item)
        
    return train_list,test_list

def list_to_str(in_list):
    out_str=""
    for i in range(len(in_list)):
        out_str = out_str + in_list[i]+','
    return out_str
```
